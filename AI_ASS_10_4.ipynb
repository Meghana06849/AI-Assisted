```python
def analyze_duplicates(numbers):
    """
    Analyze a list of numbers to identify duplicates.

    Args:
        numbers (list): A list of hashable values.

    Returns:
        tuple:
            int: The number of duplicate elements found.
            list: A list of unique elements from the input.
    """
    duplicate_count = 0
    unique_values = []

    for value in numbers:
        if value not in unique_values:
            unique_values.append(value)
        else:
            duplicate_count += 1

    return duplicate_count, unique_values
```

```python
import sqlite3

def get_user_by_id(user_id):
    """
    Retrieve a user record by user ID securely.

    Args:
        user_id (int): The unique ID of the user.

    Returns:
        tuple | None: User record if found, otherwise None.
    """
    if not isinstance(user_id, int):
        raise ValueError("user_id must be an integer")

    try:
        conn = sqlite3.connect("users.db")
        cursor = conn.cursor()

        query = "SELECT * FROM users WHERE id = ?"
        cursor.execute(query, (user_id,))

        result = cursor.fetchone()
        return result

    except sqlite3.Error as db_error:
        # Log this in real systems
        print(f"Database error: {db_error}")
        return None

    except Exception as error:
        print(f"Unexpected error: {error}")
        return None

    finally:
        if 'conn' in locals():
            conn.close()
```

```python
def analyze_duplicates(values):
    """
    Analyze a list to identify duplicate elements.

    Args:
        values (list): A list of hashable elements.

    Returns:
        tuple:
            int: Number of duplicate occurrences.
            list: List of duplicate values.
    """
    if not isinstance(values, list):
        raise TypeError("Input must be a list")

    seen = set()
    duplicates = set()
    duplicate_count = 0

    for item in values:
        if item in seen:
            duplicate_count += 1
            duplicates.add(item)
        else:
            seen.add(item)

    return duplicate_count, list(duplicates)
```
