### User-Defined Function to Calculate Energy Charges (EC)

This function takes the `units_consumed` and `customer_type` as input and applies different tariff rates based on the customer category and consumption slabs.

```python
def calculate_energy_charges(units_consumed, customer_type):
    """
    Calculates energy charges based on units consumed and customer type.

    Args:
        units_consumed (float): The total units of electricity consumed.
        customer_type (str): The type of customer (e.g., 'residential', 'commercial', 'industrial').

    Returns:
        float: The calculated energy charges.
    """
    energy_charges = 0.0

    if units_consumed < 0:
        print("Warning: Units Consumed cannot be negative. Energy charges set to 0.")
        return 0.0

    customer_type_lower = customer_type.lower()

    if customer_type_lower == 'residential' or customer_type_lower == 'domestic':
        if units_consumed <= 100:
            energy_charges = units_consumed * 5.0
        else:
            # 100 units at 5.0, remaining at 7.0
            energy_charges = (100 * 5.0) + ((units_consumed - 100) * 7.0)
    elif customer_type_lower == 'commercial':
        if units_consumed <= 200:
            energy_charges = units_consumed * 8.0
        else:
            # 200 units at 8.0, remaining at 10.0
            energy_charges = (200 * 8.0) + ((units_consumed - 200) * 10.0)
    elif customer_type_lower == 'industrial':
        # Industrial customers have a flat rate
        energy_charges = units_consumed * 12.0
    else:
        print(f"Error: Unknown customer type '{customer_type}'. Energy charges cannot be calculated.")
        return 0.0

    return energy_charges
```

### User-Defined Function to Calculate Fixed Charges (FC)

This function determines fixed charges based on the customer type.

```python
def calculate_fixed_charges(customer_type):
    """
    Calculates fixed charges based on customer type.

    Args:
        customer_type (str): The type of customer (e.g., 'residential', 'commercial', 'industrial').

    Returns:
        float: The calculated fixed charges.
    """
    fixed_charges = 0.0
    customer_type_lower = customer_type.lower()

    if customer_type_lower == 'residential' or customer_type_lower == 'domestic':
        fixed_charges = 50.0  # Example fixed charge for residential
    elif customer_type_lower == 'commercial':
        fixed_charges = 150.0 # Example fixed charge for commercial
    elif customer_type_lower == 'industrial':
        fixed_charges = 500.0 # Example fixed charge for industrial
    else:
        print(f"Error: Unknown customer type '{customer_type}'. Fixed charges set to 0.")
        return 0.0

    return fixed_charges
```

### Main Program: Reading Inputs, Calculating, and Displaying Results

This section reads the required inputs, calculates units consumed, and then uses the defined functions to determine energy and fixed charges, finally displaying a detailed bill.

```python
# Read Previous Units (PU)
previous_units = float(input("Enter Previous Units (PU): "))

# Read Current Units (CU)
current_units = float(input("Enter Current Units (CU): "))

# Read Type of Customer
customer_type = input("Enter Type of Customer (e.g., 'residential', 'commercial', 'industrial'): ")

# Calculate units consumed
units_consumed = current_units - previous_units

# Calculate Energy Charges using the function
energy_charges = calculate_energy_charges(units_consumed, customer_type)

# Calculate Fixed Charges using the function
fixed_charges = calculate_fixed_charges(customer_type)

# Calculate Total Bill
total_bill = energy_charges + fixed_charges

# Display the results
print(f"\n--- Electricity Bill Summary ---")
print(f"Previous Units (PU): {previous_units}")
print(f"Current Units (CU): {current_units}")
print(f"Units Consumed: {units_consumed}")
print(f"Type of Customer: {customer_type}")
print(f"Energy Charges (EC): {energy_charges:.2f}")
print(f"Fixed Charges (FC): {fixed_charges:.2f}")
print(f"------------------------------")
print(f"Total Bill: {total_bill:.2f}")
```

### User-Defined Function to Calculate Customer Charges (CC)

This function determines customer charges based on the customer type.

```python
def calculate_customer_charges(customer_type):
    """
    Calculates customer charges based on customer type.

    Args:
        customer_type (str): The type of customer (e.g., 'residential', 'commercial', 'industrial').

    Returns:
        float: The calculated customer charges.
    """
    customer_charges = 0.0
    customer_type_lower = customer_type.lower()

    if customer_type_lower == 'residential' or customer_type_lower == 'domestic':
        customer_charges = 20.0  # Example customer charge for residential
    elif customer_type_lower == 'commercial':
        customer_charges = 75.0 # Example customer charge for commercial
    elif customer_type_lower == 'industrial':
        customer_charges = 200.0 # Example customer charge for industrial
    else:
        print(f"Error: Unknown customer type '{customer_type}'. Customer charges set to 0.")
        return 0.0

    return customer_charges
```

### User-Defined Function to Calculate Electricity Duty (ED)

This function calculates electricity duty as a percentage of energy charges.

```python
def calculate_electricity_duty(energy_charges, duty_percentage=0.06):
    """
    Calculates electricity duty as a percentage of energy charges.

    Args:
        energy_charges (float): The calculated energy charges.
        duty_percentage (float): The percentage rate for electricity duty (e.g., 0.06 for 6%).

    Returns:
        float: The calculated electricity duty.
    """
    if energy_charges < 0:
        return 0.0 # No duty on negative charges (credit)
    return energy_charges * duty_percentage
```
