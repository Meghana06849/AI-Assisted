```python
class Calculator:
    """A simple calculator module providing basic arithmetic operations.

    This module contains functions for addition, subtraction, multiplication, and division.
    It is intended for use in various projects requiring fundamental numerical computations.

    Attributes
    ----------
    None

    Examples
    --------
    >>> calc = Calculator()
    >>> calc.add(5, 3)
    8
    >>> calc.subtract(10, 4)
    6
    """

    def add(self, a, b):
        """Add two numbers.

        Parameters
        ----------
        a : int or float
            The first number.
        b : int or float
            The second number.

        Returns
        -------
        int or float
            The sum of `a` and `b`.

        Examples
        --------
        >>> calc = Calculator()
        >>> calc.add(2, 3)
        5
        >>> calc.add(-1, 10.5)
        9.5
        """
        return a + b

    def subtract(self, a, b):
        """Subtract the second number from the first.

        Parameters
        ----------
        a : int or float
            The minuend (number from which another is subtracted).
        b : int or float
            The subtrahend (number to be subtracted).

        Returns
        -------
        int or float
            The difference between `a` and `b`.

        Examples
        --------
        >>> calc = Calculator()
        >>> calc.subtract(10, 5)
        5
        >>> calc.subtract(3.5, 1.2)
        2.3
        """
        return a - b

    def multiply(self, a, b):
        """Multiply two numbers.

        Parameters
        ----------
        a : int or float
            The first number.
        b : int or float
            The second number.

        Returns
        -------
        int or float
            The product of `a` and `b`.

        Examples
        --------
        >>> calc = Calculator()
        >>> calc.multiply(4, 6)
        24
        >>> calc.multiply(2.5, 2)
        5.0
        """
        return a * b

    def divide(self, a, b):
        """Divide the first number by the second.

        Parameters
        ----------
        a : int or float
            The dividend (number to be divided).
        b : int or float
            The divisor (number by which to divide).

        Returns
        -------
        int or float
            The quotient of `a` divided by `b`.

        Raises
        ------
        ValueError
            If the divisor `b` is zero.

        Examples
        --------
        >>> calc = Calculator()
        >>> calc.divide(10, 2)
        5.0
        >>> calc.divide(7, 0)
        Traceback (most recent call last):
            ...
        ValueError: Cannot divide by zero
        """
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b

# Example Usage:
calculator = Calculator()
print(f"5 + 3 = {calculator.add(5, 3)}")
print(f"10 - 4 = {calculator.subtract(10, 4)}")
print(f"6 * 7 = {calculator.multiply(6, 7)}")
print(f"15 / 3 = {calculator.divide(15, 3)}")
try:
    print(f"10 / 0 = {calculator.divide(10, 0)}")
except ValueError as e:
    print(e)
```

```python
class sru_student_manual_comments:
    """Represents a student in the SRU system."""

    def __init__(self, name, roll_no, hostel_status):
        # Initialize student's name
        self.name = name
        # Initialize student's roll number
        self.roll_no = roll_no
        # Initialize student's hostel status (e.g., 'Resident', 'Day Scholar')
        self.hostel_status = hostel_status
        # Initialize fee balance, default to 0
        self.fee_balance = 0

    def fee_update(self, amount, transaction_type='credit'):
        """Updates the student's fee balance."""
        if transaction_type == 'credit':
            # Decrease fee balance when payment is received
            self.fee_balance -= amount
            print(f"Fee payment of {amount} received. New balance: {self.fee_balance}")
        elif transaction_type == 'debit':
            # Increase fee balance when fees are charged
            self.fee_balance += amount
            print(f"Fee of {amount} charged. New balance: {self.fee_balance}")
        else:
            # Handle invalid transaction type
            print("Invalid transaction type. Use 'credit' or 'debit'.")

    def display_details(self):
        """Displays the student's details and current fee balance."""
        # Print student's name
        print(f"Student Name: {self.name}")
        # Print student's roll number
        print(f"Roll Number: {self.roll_no}")
        # Print student's hostel status
        print(f"Hostel Status: {self.hostel_status}")
        # Print current fee balance
        print(f"Fee Balance: {self.fee_balance}")

# Example Usage for manual comments version:
manual_student = sru_student_manual_comments("Manual User", "SRU_M001", "Resident")
manual_student.display_details()
manual_student.fee_update(1000, 'debit')
manual_student.fee_update(500, 'credit')
manual_student.display_details()
```

```python
def sum_even_odd(numbers):
    """Calculates the sum of even and odd numbers in a list.

    This function iterates through a list of integers and separates them
    into even and odd numbers, then calculates the sum of each group.

    Args:
        numbers (list[int]): A list of integer numbers.

    Returns:
        tuple[int, int]: A tuple containing two integers:
                         - The sum of even numbers.
                         - The sum of odd numbers.

    Examples:
        >>> sum_even_odd([1, 2, 3, 4, 5, 6])
        (12, 9)
        >>> sum_even_odd([])
        (0, 0)
        >>> sum_even_odd([7, 9, 11])
        (0, 27)
    """
    sum_even = 0
    sum_odd = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
        else:
            sum_odd += num
    return sum_even, sum_odd

# Execute the function and display results
even_sum, odd_sum = sum_even_odd([1, 2, 3, 4, 5, 6])
print(f"Function Output: Sum of even numbers: {even_sum}, Sum of odd numbers: {odd_sum}\n")
```
